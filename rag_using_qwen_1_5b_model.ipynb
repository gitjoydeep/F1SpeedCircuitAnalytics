```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-generation", model="Qwen/Qwen2.5-1.5B-Instruct")

```

```python
question1 = "what is ACE ?"
question2 = "what is volley"
```

```python
prompt = "Answer the questions \n" + \
         "Question: " + question1 + "\n" + \
         "Answer: "

print(prompt)
```

```python
results = pipe(prompt, max_new_tokens = 50)
print(results)
```
